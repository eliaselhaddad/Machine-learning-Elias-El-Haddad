```python
# We want to simulate data  (x,y)  to represent cost for phone subscriptions, with:

# x  - called minutes per month
# y  - SEK per month
#   a) Use numpy.random.normal() to simulate a dataset with the following requirements:(*)

# set a seed to 42 (for reproducibility and reference)
# simulate 400 x-values from the r.v.  X∼N(100,100) 
# take absolute value of these x-values
# simulate noise 400 noise values from r.v.  ϵ∼N(0,50) 
# Let  y=2x+25+ϵ 
# plot the data set
#   b) Now we want to remove some outliers according to this assumption: (*)

# no one talks more than 300 min using this type of subscription
# no ones costs can be negative
# plot the new dataset
# also plot ground truth using the true parameters  β0=25,β1=2 
#   c) Insert the values into a DataFrame (*)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

np.random.seed(42)

# Simulate x-values
x = np.abs(np.random.normal(100, 100, 400))

# Simulate noise
eps = np.random.normal(0, 50, 400)

# Calculate y-values
y = 2*x + 25 + eps

# Plot the dataset
plt.scatter(x, y)
plt.xlabel('Minutes')
plt.ylabel('Cost (SEK)')
plt.title('Simulated Phone Dataset')
plt.show()

# Remove outliers
x_filtered = x[(x <= 300) & (y >= 0)]
y_filtered = y[(x <= 300) & (y >= 0)]

# Plot the filtered dataset
plt.scatter(x_filtered, y_filtered)
plt.plot(x, 2*x + 25, color='red')
plt.xlabel('Minutes')
plt.ylabel('Cost (SEK)')
plt.title('Filtered Phone Dataset')
plt.show()

# Create a DataFrame
df = pd.DataFrame({'Minutes': x_filtered, 'Cost': y_filtered})

# Print the first few rows of the DataFrame
print(df.head())
```
