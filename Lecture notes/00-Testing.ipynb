```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

x = np.abs(np.random.normal(100, 100, 400))
eps = np.random.normal(0, 50, 400)
y = 2 * x + 25 + eps

plt.scatter(x, y)
plt.xlabel('Minutes per month')
plt.ylabel('Cost per month (SEK)')
plt.show()
```

```python
mask = (x <= 300) & (y >= 0) 
x_filtered = x[mask]
y_filtered = y[mask]

x_filtered
```

```python
plt.scatter(x_filtered, y_filtered, label='Filtered Data')
plt.plot(x, 2*x+25, 'r', label='Ground Truth')
plt.xlabel('Minutes per month')
plt.ylabel('Cost per month (SEK)')
plt.legend()
plt.show()
```

```python
data = pd.DataFrame({'Minutes': x_filtered, 'Cost': y_filtered})
print(data.head())
```
